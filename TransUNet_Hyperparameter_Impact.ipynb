### Test for diffrent number of epochs

```python
# Common parameters
common_params = "--dataset Synapse --vit_name R50-ViT-B_16 --batch_size 6 "

epochs_to_test = [5, 10, 15, 25, 50, 100, 150]

for nepoch in epochs_to_test:
    # Run the training script
    print("Train started for number of epoch: ", nepoch)
    train_command = f"CUDA_VISIBLE_DEVICES=0 python train.py {common_params} --max_epochs {nepoch}"
    !{train_command}

    # Run the testing script and capture the metrics
    print("Test started for number of epoch: ", nepoch)
    test_command = f"python test2.py {common_params} --max_epochs {nepoch}"
    !{test_command}
```
